# Basic Car Speed 

## Problem Statement

We would like to determine the speed of a car based on the bumpiness and slope of the road. For this, we have a list of the speed from a list of drivers.

```python
import pandas as pd
```

```python
%pylab inline
```

```python
df = pd.read_csv("terrain.csv")
```

```python
df[["bumpiness","grade"]]
```

```python
X = df[["bumpiness","grade"]].values
y = df["target"]
```

```python
def draw(clf,X,y,h=0.05):
    
    plt.figure(figsize=(10,10))

    x_min, x_max = X[:, 0].min() - 0.05, X[:, 0].max() + .05
    y_min, y_max = X[:, 1].min() - .05, X[:, 1].max() + .05
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    
    if clf is not None:
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        Z = Z.reshape(xx.shape)
        plt.scatter(xx,yy,c=Z)
    else:
        plt.scatter(xx,yy)
    
    plt.scatter(X[:,0],X[:,1],c=y,cmap="Paired")
```

```python
# Load the library
from sklearn.linear_model import LogisticRegression
# Create an instance of the classifier
clf=LogisticRegression()
# Fit the data
clf.fit(X,y)
```

```python
plt.scatter(df["bumpiness"],df["grade"],c=df["target"]);
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1)
```

```python
clf.fit(X_train,y_train)
```

```python
cross_val_score(clf,X,y,cv=5,scoring="accuracy").mean()
```

```python
draw(clf,X,y,0.01)
```

```python
draw(clf,X_train,y_train,0.01)
```

```python
draw(clf,X_test,y_test,0.01)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
knGrid = GridSearchCV(clfk,cv=5,param_grid={'n_neighbors':np.arange(20,50)})
```

```python
knGrid.fit(X,y)
```

```python
knGrid.best_params_
```

# Decision Tree Classifier

```python
# Import library
from sklearn.tree import DecisionTreeClassifier
# Create instance
clfTree = DecisionTreeClassifier(min_samples_leaf=20,max_depth=3)
# Fit the data
clfTree.fit(X,y)
```

```python
draw(clfTree,X,y,0.01)
```
